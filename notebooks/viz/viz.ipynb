# Visualization

Let's run through the various types of visualizations that not-again-ai provides.


## Bar Plots

Saves a simple barplot to the specified pathname given a list of values for the x-axis and a list of values for the y-axis. This function calls [seaborn.barplot](https://seaborn.pydata.org/generated/seaborn.barplot.html).

```python
from pathlib import Path

from PIL import Image

from not_again_ai.viz.barplots import simple_barplot

x = ["fence", "wall", "gate", "door", "window", "counter", "stair", "curtain", "ceiling", "floor"]
y = [0.5, 0.01, 0.25, 0.2, 0.1, 0.05, 0.04, 0.03, 0.02, 0.3]

simple_barplot(
    x,
    y,
    Path("barplot.png"),
    order="asc",
    title="Token logits",
    x_label="token",
    y_label="logit",
)

Image.open("barplot.png")
```

## Distributions

Saves a univariate distribution plot to the specified pathname given a list of numbers. This function calls [seaborn.displot](https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot).

```python
import numpy as np

from not_again_ai.viz.distributions import univariate_distplot

univariate_distplot(
    data=np.random.normal(size=1000),
    save_pathname="distribution.png",
    print_summary=False,
    title="Normal Distribution",
    xlabel="Test xlabel",
    ylabel="Test ylabel",
    xlim=(-5, 5),
    ylim=(0, 40),
    xticks=None,
    yticks=None,
    bins=200,
    font_size=54,
    height=14,
    aspect=2.2,
)

Image.open("distribution.png")
```

## Scatter Plots

Saves a basic scatterplot to the specified pathname given corresponding x and y values as lists. This function leverages [seaborn's scatterplot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html).

```python
from not_again_ai.viz.scatterplot import scatterplot_basic

rs = np.random.RandomState(365)
x = rs.randn(100)
e = rs.randn(100) * 0.2
y = x + e
scatterplot_basic(
    x,
    y,
    save_pathname="scatterplot.png",
    title=None,
    xlim=(-5, 5),
    ylim=(-5, 5),
    font_size=36,
    height=15,
    aspect=1.8,
)

Image.open("scatterplot.png")
```

## Time Series Plots

Saves a time series plot where each row in `ts_data` is a time series.

```python
import matplotlib
import pandas as pd

from not_again_ai.viz.time_series import ts_lineplot

rs = np.random.RandomState(365)
values = rs.randn(365, 4).cumsum(axis=0).T
dates = pd.date_range("1 1 2024", periods=365, freq="D")

ts_lineplot(
    ts_data=values,
    save_pathname="ts.png",
    ts_x=dates,
    ts_names=["A", "B", "C", "D"],
    title="Example Time Series",
    xlabel=None,
    ylabel=None,
    yticks=np.arange(-30, 40, 10),
    xaxis_date_format="%b '%y",
    xaxis_major_locator=matplotlib.dates.MonthLocator((1, 3, 5, 7, 9, 11)),
    font_size=46,
    linewidth=1.8,
    legend_loc=2,
    palette="colorblind",
)

Image.open("ts.png")
```
